Using cdo to do select the data, calculate the anomaly and matplotlib to display it.

```python
from netCDF4 import Dataset as NetCDFFile
import matplotlib.pyplot as plt
import sys, datetime, numpy
```

NorESM2-MMtas.WA.2x2b.30.nc is NorESM2-MM tas 1980-2010 west africa region monthly regridded to 2 degrees - created using cdo.

```python
file_location = '/gws/pw/j05/cop26_hackathons/leeds/ministry/Tom_W/afr_T/NorESM2-MMtas.WA.2x2b.30.nc'
nc = NetCDFFile(file_location)
TA = nc.variables['temperature_anomaly']
TA = numpy.mean(TA,axis=(1,2))
```

```python
plt.figure(figsize=(12,6))
x=numpy.linspace(1980,2010,360)
plt.plot(x,TA, label='NorESM2-MM', alpha=0.8)
plt.title('CMIP6 2m NorESM2-MM Temperature Anomaly West Africa 1980-2010')
plt.xlabel('Time')
plt.ylabel('Temperature Anomaly (degrees K)')
plt.xlim(1980,2010)
plt.legend()
plt.show()
```
